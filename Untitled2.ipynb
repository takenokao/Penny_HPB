```python
#到着年ごとの人数の順位('number'の列は表示しない)
result = df.groupby('datearr').size().reset_index()
result.columns = ['datearr','number']
result['rank'] = result['number'].rank(ascending = True, method = 'min')
result.drop('number',axis = 1,inplace = True)
result
```

```python
#到着年ごとの人数の順位('number'を表示する)
result = df.groupby(['datearr']).size().reset_index()
result.columns = ['datearr','number']
result['rank'] = result['number'].rank(ascending = False)
result
```

```python
#rankが順番になるように並び替える並び替える
result2 = result.sort_values(by='rank')
result2
```

```python
# 集約②++ 到着年ごとの人数の順位（全ての値を整数値に変換）
result2.astype(int)
```

```python
#到着年ごとの年齢の最大値
result = df.groupby('datearr').agg({'age':['max']}).reset_index()
result.columns = ['datearr','age_max']
result
```

```python
#到着年ごとの年齢の分散値と標準偏差
result = df.groupby('datearr').agg({'age':['var','std']}).reset_index()
result.columns = ['datearr','age_var','age_std']
result
```
